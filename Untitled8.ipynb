```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/tec03/Datasets/main/datasets/antropometria-dataset.csv')
```

```python
data.head()
```

```python
variables = ['peso', 'cintura']

for var in variables:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='sexo', y=var, data=data)
    plt.title(f'Box Plot de {var} por Sexo')
    plt.xlabel('Sexo')
    plt.ylabel(var.capitalize())
    plt.show()
```

```python
for var in variables:
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='sexo', y=var, data=data, marker='o')
    plt.title(f'Line Plot de {var} por Sexo')
    plt.xlabel('Sexo')
    plt.ylabel(var.capitalize())
    plt.show()
```

```python
for var in variables:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x=var, hue='sexo', kde=True)
    plt.title(f'Histograma de {var} por Sexo')
    plt.xlabel(var.capitalize())
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='peso', y='cintura', hue='sexo', data=data)
plt.title('Diagrama de DispersiÃ³n entre Peso y Cintura por Sexo')
plt.xlabel('Peso')
plt.ylabel('Cintura')
plt.legend(title='Sexo', loc='upper left')
plt.show()
```

```python
data_sexo_1 = data[data['sexo'] == 1]
data_sexo_2 = data[data['sexo'] == 2]
```

```python
for var in variables:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data_sexo_1, x=var, kde=True, color='blue')
    plt.title(f'Histograma de {var} para Sexo 1')
    plt.xlabel(var.capitalize())
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
for var in variables:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data_sexo_2, x=var, kde=True, color='red')
    plt.title(f'Histograma de {var} para Sexo 2')
    plt.xlabel(var.capitalize())
    plt.ylabel('Frecuencia')
    plt.show()
```
